# Seaborn swarmplot

## 1. Importing modules and dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

Left only cars with even number of cylinders

```python
cars = cars[cars['cylinders'].isin([4,6,8])]
```

## 1.2 Data visualizations

### Basics

```python
sns.swarmplot(cars['horsepower']);
```

```python
sns.swarmplot(x=cars['origin'], y=cars['horsepower']);
```

There's another syntax for plot above, when all columns you wanna plot are in the same dataframe:

```python
sns.swarmplot(data=cars, x='origin', y='horsepower');
```

Adding hue:

```python
sns.swarmplot(data=cars, x='origin', y='horsepower', hue='cylinders');
```

dodge argument:

```python
sns.swarmplot(data=cars, x='origin', y='horsepower', hue='cylinders', 
              dodge=True);
```

dodge splits all of the three these groups horizontaly
